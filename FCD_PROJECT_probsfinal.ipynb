### 2.3 Convert dates

```python
def parse_mixed_date(x):
    try:
        # excel serial date
        x_float = float(x)
        return pd.Timestamp('1899-12-30') + pd.to_timedelta(x_float, unit='D')
    except:
        try:
            # DD/MM/YY
            return pd.to_datetime(x, format="%d/%m/%y", dayfirst=True, errors="coerce")
        except:
            return pd.NaT

df["DN"] = df["DN"].apply(parse_mixed_date)
df["ATENDIMENTO"] = df["ATENDIMENTO"].apply(parse_mixed_date)
```

The dates were converted to a consistent datetime format to handle mixed representations, including Excel serial numbers and string formats. This ensures a consistent dataset, and prevents errors from inconsistent or misinterpreted date values.

### 2.4 Filter numeric variables and complete missing values

```python
def repair_row(row):
    PESO, ALTURA, IMC = row["PESO"], row["ALTURA"], row["IMC"]
    missing = sum(pd.isna([PESO, ALTURA, IMC]))

    if missing >= 2:
        return row

    if pd.isna(PESO):
        row["PESO"] = IMC * (ALTURA ** 2)
    elif pd.isna(ALTURA) and IMC > 0:
        row["ALTURA"] = np.sqrt(PESO / IMC)
    elif pd.isna(IMC):
        row["IMC"] = PESO / (ALTURA ** 2)

    return row

# IDADE  ---  (2 < idade < 19)
df["IDADE"] = pd.to_numeric(df["IDADE"], errors="coerce")
df = df[df["IDADE"].between(2, 19)]

numeric_cols = ["PESO", "ALTURA", "IMC", "PA_SISTOLICA", "PA_DIASTOLICA", "FC"]
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors="coerce")

# PESO  ---  (8 < PESO < 100)
df.loc[df["PESO"] <= 0, "PESO"] = np.nan

# ALTURA  ---  ajustes + conversão + limites (60cm < ALTURA < 220cm)
df.loc[df["ALTURA"] == 0, "ALTURA"] = np.nan
df.loc[df["ALTURA"] < 3, "ALTURA"] = df["ALTURA"]           # já está em metros
df.loc[df["ALTURA"] > 100, "ALTURA"] = df["ALTURA"] / 100  # estava em cm

df = df.apply(repair_row, axis=1)

# ajustar novamente após repair
df.loc[df["ALTURA"] == 0, "ALTURA"] = np.nan
df.loc[df["ALTURA"] < 3, "ALTURA"] = df["ALTURA"]
df.loc[df["ALTURA"] > 100, "ALTURA"] = df["ALTURA"] / 100

# aplicar limites finais
df.loc[~df["PESO"].between(8, 100), "PESO"] = np.nan
df.loc[~df["ALTURA"].between(0.60, 2.20), "ALTURA"] = np.nan
df.loc[~df["IMC"].between(10, 60), "IMC"] = np.nan

# PA_SISTOLICA  --- 50 < PAS < 250
df["PA_SISTOLICA"] = df["PA_SISTOLICA"].clip(lower=50, upper=250)

# PA_DIASTOLICA  --- 30 < PAD < 150
df["PA_DIASTOLICA"] = df["PA_DIASTOLICA"].clip(lower=30, upper=150)

# Definir limites razoáveis para crianças/adolescentes
df.loc[(df["FC"] < 30) | (df["FC"] > 190), "FC"] = np.nan
```

Main Changes Needed in the Raw Data:

* **PESO** There are negative values and others unreasonably high concerning in the PESO attribute, all of which are prejudicial to the proper analysis of the dataset. In order to solve this inconsistency, as a first step, the PESO values with absurd values will be refactured using the BMI and Height, if either of these values are missing or the resulting Weight calculated is unreasonable the PESO attribute will be set to Nan.

* **ALTURA** The height, as it is, is also problematic, even more than PESO. The abundancy of near zero values, due to the data pertaining mostly to newborns, is unreliable and skews the whole ALTURA column towards 0. As a rule of thumb, patients with less than 2 years will be desconsidered for the data analysis and the same refactoring and discarding approach used in PESO will be taken.

* **IMC** BMIs that do not fall within the range of 10 and 60 will be discarded, as they are unrealistic. The refactoring proccess used previously will be also applied in IMC.

* **PA SISTOLICA and PA DIASTOLICA** Taking into account normal blood pressure levels for boys and girls from the age of 2 to 18 only values within the following ranges will be used for PA SISTOLICA and PA DIASTOLICA, respectively ,50 < PAS < 250 30 < PAD < 150.

* **FC:** FC was converted to numeric to ensure it can be treated quantitatively in analyses. Values outside a physiologically reasonable range for children and adolescents (<30 or >190 bpm) were set to NaN to remove outliers and avoid distorting results.

Further changes:

* **AGE** Ages lower than two are already beind disconsidered, so taking into account the target scope of the analysis, that is, kids and teens, patient entries with age greater than 19 will also be discarded.

* **Unit Inconsistencies Check** The range of ALTURA 0 to 198  and PESO –40 to 157 suggests mixed measurement units, for example, cm vs. m, kg vs. g. To ensure data consistency, all measurements will use the same units and will be converted if necessary.

* **Verify Data Type Consistency** Make sure all numeric columns are actually numeric `float` or `int`.  sometimes Excel imports numbers as strings with commas or spaces.

### 2.5 Standardize categorical variables

```python
#---PPA---
df["PPA"] = df["PPA"].replace("Não Calculado", np.nan)

df["PPA"] = df["PPA"].str.strip()

# mapa de uniformização
ppa_map = {
    "Normal": "Normal",
    "Pre-Hipertensão PAS": "Pre-Hipertensão",
    "Pre-Hipertensão PAD": "Pre-Hipertensão",
    "HAS-1 PAS": "HAS-1",
    "HAS-1 PAD": "HAS-1",
    "HAS-2 PAS": "HAS-2",
    "HAS-2 PAD": "HAS-2",
    "Não Calculado": np.nan
}

df["PPA"] = df["PPA"].replace(ppa_map)

# definir ordem categórica
ppa_order = ["Normal", "Pre-Hipertensão", "HAS-1", "HAS-2"]
df["PPA"] = pd.Categorical(df["PPA"], categories=ppa_order, ordered=True)

df["PPA_BIN"] = df["PPA"].map({
    "Normal": 0,
    "Pre-Hipertensão": 1,
    "HAS-1": 1,
    "HAS-2": 1
}).astype("float")


#---NORMAL_X_ANORMAL---
df["NORMAL_X_ANORMAL"] = df["NORMAL_X_ANORMAL"].map({
    "Normal": "Normal",
    "Normais": "Normal",
    "anormal": "Anormal",
    "Anormal": "Anormal"
})

df["NORMAL_X_ANORMAL"] = df["NORMAL_X_ANORMAL"].str.capitalize()

df["NORMAL_X_ANORMAL"] = pd.Categorical(
    df["NORMAL_X_ANORMAL"],
    categories=["Normal", "Anormal"],
    ordered=True
)

df["NORMAL_X_ANORMAL_BIN"] = df["NORMAL_X_ANORMAL"].map({
    "Normal": 0,
    "Anormal": 1
}).astype("float")

#---B2---
df["B2"] = (
    df["B2"].map({
        "Normal": "Normal",
        "Hiperfonética": "Alterada",
        "Desdob fixo": "Alterada",
        "Única": "Alterada",
        "Outro": "Alterada"
    })
).astype("category")

df["B2_BIN"] = df["B2"].map({
    "Normal": 0,
    "Alterada": 1
}).astype("float")

#---SOPRO---
df["SOPRO"] = df["SOPRO"].map({
    "ausente": "Normal",
    "Sistólico": "Alterado",
    "sistólico": "Alterado",
    "diastólico": "Alterado",
    "contínuo": "Alterado",
    "Contínuo": "Alterado",
    "Sistolico e diastólico": "Alterado"
}).astype("category")

df["SOPRO_BIN"] = df["SOPRO"].map({
    "Normal": 0,
    "Alterado": 1
}).astype("float")

#---HDA---
df["HDA"] = df["HDA"].astype("category")
df["HDA_CAT"] = df["HDA"].apply(lambda x: "Assintomático" if x == "Assintomático" else ("Sintomático" if pd.notna(x) else np.nan)).astype("category")
df["HDA_BIN"] = df["HDA"].apply(lambda x: 0 if x == "Assintomático" else 1 if pd.notna(x) else np.nan).astype("float")

#---SEXO---
df["SEXO"] = df["SEXO"].replace({
    "Masculino": "M",
    "Feminino": "F",
    "masculino": "M"
}).astype("category")

df["SEXO"] = df["SEXO"].where(df["SEXO"].isin(["M", "F"]), np.nan)

df["SEXO_BIN"] = df["SEXO"].map({"F":0, "M":1}).astype("float")

#---MOTIVO1---
df["MOTIVO1"] = df["MOTIVO1"].astype("category")

df["MOTIVO1_BIN"] = df["MOTIVO1"].map({
    "5 - Parecer cardiológico": 1,
    "6 - Suspeita de cardiopatia": 1,
    "1 - Cardiopatia já estabelecida": 1,
    "2 - Check-up": 0,
    "7 - Outro": 0,
    np.nan: np.nan
}).astype("float")

#---MOTIVO2---
counts = df["MOTIVO2"].value_counts()
rare_categories = counts[counts < 200].index
df["MOTIVO2"] = df["MOTIVO2"].replace(rare_categories, "Outro").astype("category")

df["MOTIVO2_BIN"] = df["MOTIVO2"].apply(lambda x: 0 if x=="Outro" else 1 if pd.notna(x) else np.nan).astype("float")
```

**PPA:** Values labeled as “Não Calculado” were replaced with NaN because they do not provide usable information. Existing categories were mapped to standardized groups (“Normal”, “Pre-Hypertension”, “HAS-1”, “HAS-2”) for consistency and to reduce redundancy. An ordered categorical type was defined to preserve the clinical severity ranking.

**NORMAL_X_ANORMAL:** Misspellings and plural forms were corrected to unify categories as “Normal” or “Anormal”. A categorical variable with order was created for analysis, and a binary version (0 = Normal, 1 = Anormal) was generated for modeling purposes.

**B2:** Rare or variant labels were grouped into “Alterada” while preserving “Normal” to simplify the feature into a binary-like interpretation of normal vs altered. This reduces noise from low-frequency categories.

**SOPRO:** Different spellings and forms of murmurs were standardized into “Normal” (absent) or “Alterado” (any presence of murmur) to make the variable consistent and suitable for modeling.

**HDA:** Transformed into a binary variable representing symptom presence: “Assintomático” vs “Sintomático” (and also 0/1 numeric). This simplifies complex free-text data into a usable feature for analysis.

**SEXO:** Variations in gender labels were standardized to “M” and “F”. Any ambiguous or rare values were converted to NaN to prevent misclassification.

**MOTIVO2:** Rare categories (frequency < 200) were grouped into “Outro” to reduce sparsity and make the variable more robust for statistical and machine learning analysis.

### Regression

## 3.2 NUMERICAL X CATEGORICAL

In this section, we'll apply either the t-test or ANOVA.

The t-test compares the means of a numeric variable between two groups to see if they differ significantly.

ANOVA extends this idea to more than two groups, testing whether at least one group mean differs from the others. Both rely on p-values for significance, but ANOVA is necessary when there are multiple categories.

```python
# Separar colunas binárias e multiclasse
binary_cols = [col for col in categorical_cols if df[col].nunique(dropna=True) == 2]
multi_cols = [col for col in categorical_cols if df[col].nunique(dropna=True) > 2]

# ---Binário: t-test---
binary_results = []

for cat_col in binary_cols:
    df_bin = df[df[cat_col].notna()]

    groups = df_bin[cat_col].unique()
    if len(groups) != 2:
        continue

    for num_col in numeric_cols:
        group1 = df_bin[df_bin[cat_col]==groups[0]][num_col].dropna()
        group2 = df_bin[df_bin[cat_col]==groups[1]][num_col].dropna()

        if len(group1) < 2 or len(group2) < 2:
            continue

        t_stat, p_val = ttest_ind(group1, group2)
        diff_mean = group1.mean() - group2.mean()

        binary_results.append({
            "Categorical": cat_col,
            "Numeric": num_col,
            "Group1": groups[0],
            "Group2": groups[1],
            "Mean_Diff": diff_mean,
            "t_stat": t_stat,
            "p_value": p_val
        })

binary_df = pd.DataFrame(binary_results).sort_values("p_value")
print("=== Binário (t-test) ===")
print(binary_df.head(10))

# ---multiclasse: ANOVA---
multi_results = []

for cat_col in multi_cols:
    df_multi = df[df[cat_col].notna()]

    for num_col in numeric_cols:
        groups = [df_multi[df_multi[cat_col]==val][num_col].dropna() for val in df_multi[cat_col].unique()]
        if len(groups) < 2:
            continue
        f_stat, p_val = f_oneway(*groups)
        means = [g.mean() for g in groups]

        multi_results.append({
            "Categorical": cat_col,
            "Numeric": num_col,
            "Means_by_Group": means,
            "F_stat": f_stat,
            "p_value": p_val
        })

multi_df = pd.DataFrame(multi_results).sort_values("p_value")
print("\n Multiclasse (ANOVA):")
print(multi_df.head(10))
```

**Interpretation:** The t-test results for binary variables show that heart rate (FC) is significantly higher in the Anormal group compared to Normal, indicating a physiological difference.

Weight (PESO) and age (IDADE) are lower in Sintomático compared to Assintomático (HDA_CAT), highlighting that symptomatic patients tend to be younger and lighter.

## 3.3 CATEGORICA X CATEGORICA

The categorical x categorical analysis examines the relationships between categorical variables using contingency tables, Chi-squared tests, and measures of association like Cramér's V. This approach identifies which variables are strongly linked and may be important for predicting outcomes. We will use it to highlight the most informative categorical features in the dataset.

```python
# Seleciona apenas colunas categóricas relevantes (ou converte colunas para 'category')
categorical_cols = ['NORMAL_X_ANORMAL', 'B2', 'SOPRO', 'HDA', 'SEXO', 'MOTIVO1']  # ajustar conforme necessário
df[categorical_cols] = df[categorical_cols].astype('category')

# Função para calcular Cramér's V
def cramers_v(confusion_matrix):
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    min_dim = min(confusion_matrix.shape) - 1
    return np.sqrt(chi2 / (n * min_dim))

# Exemplo: comparar duas variáveis categóricas
var1 = 'NORMAL_X_ANORMAL'
var2 = 'SOPRO'

# Criar tabela de contingência
contingency = pd.crosstab(df[var1], df[var2])
print("=== Contingency Table ===")
print(contingency)

# Teste qui-quadrado
chi2, p, dof, expected = chi2_contingency(contingency)
print("\n=== Chi-squared Test ===")
print(f"Chi2 Statistic: {chi2:.3f}")
print(f"P-value: {p:.3e}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
print(pd.DataFrame(expected, index=contingency.index, columns=contingency.columns))

# Medida de associação: Cramér's V
v = cramers_v(contingency)
print(f"\nCramér's V: {v:.3f}")

# Para comparar todas as colunas categóricas de uma vez
print("\n=== Pairwise Categorical Associations (Cramér's V) ===")
results = pd.DataFrame(index=categorical_cols, columns=categorical_cols, dtype=float)
for col1 in categorical_cols:
    for col2 in categorical_cols:
        if col1 == col2:
            results.loc[col1, col2] = 1.0
        else:
            contingency = pd.crosstab(df[col1], df[col2])
            results.loc[col1, col2] = cramers_v(contingency)

print(results.round(2))
```

**Interpretation:** The results make sense and reveal a very strong association between NORMAL_X_ANORMAL and SOPRO, as shown by the high Chi-squared statistic and extremely small p-value (p < 0.001).

The observed counts differ substantially from the expected counts, indicating that abnormal heart sounds are much more common in the "Anormal" group. The Cramér's V of 0.85 confirms a very strong relationship, making SOPRO the most predictive categorical variable for distinguishing normal versus abnormal cases. Other associations are much weaker, with B2 and MOTIVO1 showing moderate links (Cramér's V ~0.2-0.3), and SEXO and HDA being largely independent from NORMAL_X_ANORMAL. Overall, the analysis highlights SOPRO as the key categorical feature for predicting cardiac abnormalities.

## 5.2 Support Vector Machines

Support Vector Machines (SVM) are used to classify patients based on clinical features. The algorithm identifies the optimal hyperplane that separates the target classes, allowing prediction of cardiac outcomes. In this project, we're going to be using different groups of variables to train an SVM model and test its accuracy, in order to determine which variables are most useful for identifying cardiovascular diseases in children.

**GROUP 0:** PESO, ALTURA, IMC, IDADE, PA_SISTOLICA, PA_DIASTOLICA, FC, SEXO_BIN, SOPRO_BIN, B2_BIN, HDS_BIN

We start by using a large number of variables to see what maximum accuracy can be achieved. It serves as a reference model to compare the contribution of each future group.

The code below builds a supervised learning pipeline to predict the Normal vs. Abnormal outcome using a Support Vector Machine. Missing values in the target are removed, while missing predictor values are imputed using the median to ensure a clean dataset. Features are standardized and the data is split into training and test sets with class stratification. The SVM model is then trained and evaluated using standard classification metrics.

```python
# Seleção de variáveis
target = "NORMAL_X_ANORMAL_BIN"

features = [
    "PESO",
    "ALTURA",
    "IMC",
    "IDADE",
    "PA_SISTOLICA",
    "PA_DIASTOLICA",
    "FC",
    "SEXO_BIN",
    "SOPRO_BIN",
    "B2_BIN",
    "HDA_BIN"
]

df_model = df[features + [target]].copy()

# Remover linhas com NaN no alvo
df_model = df_model.dropna(subset=[target])

# Tratamento de NaN nas features
for col in features:
    df_model[col] = df_model[col].fillna(df_model[col].median())

# Separar X e y
X = df_model[features]
y = df_model[target]

assert y.isna().sum() == 0

# Train / Test split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y
)

# Pipeline: Scaling + SVM
pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("svm", SVC(
        kernel="rbf",
        C=1.0,
        gamma="scale",
        class_weight="balanced",
        random_state=42
    ))
])

# Treino
pipeline.fit(X_train, y_train)

# Previsões
y_pred = pipeline.predict(X_test)

# Avaliação
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

Results: The model achieves high overall accuracy (~93%), indicating strong discriminative performance. The confusion matrix shows very good detection of normal cases and a slightly lower, but still solid, detection of abnormal cases. Recall for the abnormal class (0.86) indicates that most clinically relevant abnormal patients are correctly identified, although some false negatives remain. Precision is high for both classes, suggesting few false positives. Overall, the model is well balanced and clinically reliable, with a mild tendency to miss a small proportion of abnormal cases.

5. **Bayesian Network Analysis**
